## Building Machine Learning Model

```python
df = pd.read_csv('../dataset/final_dataset.csv')
```

```python
df = df.dropna()
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['spam'], test_size=0.25, random_state=42)
```

```python
tf_idf = TfidfVectorizer(stop_words='english', max_features=1000)
X_train_tfidf = tf_idf.fit_transform(X_train)
X_test_tfidf = tf_idf.fit_transform(X_test)
```

```python
dump(tf_idf, 'svm_vectorizer_spam.joblib')
```

```python
param_dist = {
    'C': [0.1, 1, 10, 100],
    'kernel': ['linear', 'rbf', 'poly'],
    'gamma': ['scale', 'auto'],
    'degree': randint(3, 6),  # Random integer between 3 and 5
    'shrinking': [True, False],
    'tol': [1e-3, 1e-4],
}
```

```python
svc = SVC()

random_search = RandomizedSearchCV(svc, param_distributions=param_dist, n_iter=50, cv=5, n_jobs=-1)
random_search.fit(X_train_tfidf, y_train)

print("Best parameters found: ", random_search.best_params_)
```

```python
svm_model = SVC(C=100, degree=4, gamma='scale', kernel='rbf', shrinking=True, tol=0.001, probability=True)
svm_model.fit(X_train_tfidf, y_train)
```

```python
y_pred = svm_model.predict(X_test_tfidf)
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', cbar=False, 
            xticklabels=['Not Spam', 'Spam'], yticklabels=['Not Spam', 'Spam'])

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix (Color-Coded)')
plt.show()
```

```python
dump(svm_model, 'svm_model.joblib')
```
